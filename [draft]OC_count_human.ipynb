## prepareing function to output sounds

```python
import wave
import pyaudio

def play_wav_file():
    # Open the WAV file
#     wav_file = wave.open('se_30103.wav', 'rb')
    wav_file = wave.open('se_30306.wav', 'rb')


    # Initialize PyAudio
    audio = pyaudio.PyAudio()

    # Open a new stream
    stream = audio.open(format=audio.get_format_from_width(wav_file.getsampwidth()),
                        channels=wav_file.getnchannels(),
                        rate=wav_file.getframerate(),
                        output=True)

    # Read data from the file and play it
    chunk_size = 1024
    data = wav_file.readframes(chunk_size)

    while data:
        stream.write(data)
        data = wav_file.readframes(chunk_size)

    # Cleanup
    stream.stop_stream()
    stream.close()
    audio.terminate()
    wav_file.close()

# Example usage
play_wav_file()
```

## Running

```python
# 3、利用者数記録

import cv2
import numpy as np
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision
from mediapipe import ImageFormat
import threading
import time
import datetime


base_options = python.BaseOptions(model_asset_path='efficientdet_lite0.tflite')
options = vision.ObjectDetectorOptions(base_options=base_options,\
                                       score_threshold=0.5)
detector = vision.ObjectDetector.create_from_options(options)
thread = threading.Thread(target=play_wav_file)

log_file = open("saved_log.txt", "a")

MARGIN = 10  # pixels
ROW_SIZE = 10  # pixels
FONT_SIZE = 1
FONT_THICKNESS = 1
TEXT_COLOR = (0, 255, 0)  # b g r
TEXT_COLOR2 = (0, 0, 255)

ssum = 0
num_bef = 0

num_before = 0

while cap.isOpened():
    success, bgr = cap.read()
    if not success:
        continue

    rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
    rgb_mp = mp.Image(image_format=ImageFormat.SRGB, data=rgb)

    detection_result = detector.detect(rgb_mp)

    num = 0

    for detection in detection_result.detections:

        category = detection.categories[0]
        category_name = category.category_name

        # Draw bounding_box
        bbox = detection.bounding_box
        start_point = bbox.origin_x, bbox.origin_y
        end_point = bbox.origin_x + bbox.width, bbox.origin_y + bbox.height
        if category_name == "person":
            cv2.rectangle(bgr, start_point, end_point, TEXT_COLOR, 3)

        if category_name == "person":
            num += 1

    if num > num_bef:
        ssum = ssum + (num - num_bef)

    if num_before < ssum:
        thread.start()
        thread = threading.Thread(target=play_wav_file)
        print(datetime.datetime.now())
        log_file.write(str(datetime.datetime.now())+"\n")
    num_before = ssum
    num_bef = num


    text_location2 = (520, 30)
    cv2.putText(bgr, "times", text_location2, cv2.FONT_HERSHEY_PLAIN,
                        2, TEXT_COLOR2, 2)

    text_location3 = (530, 100)
    cv2.putText(bgr, str(ssum), text_location3, cv2.FONT_HERSHEY_PLAIN,
                        5, TEXT_COLOR2, 3)
    bgr = cv2.resize(bgr, (640*2, 480*2))

    cv2.imshow('MediaPipe Object detection', bgr)
    key = cv2.waitKey(5)
    if key & 0xFF == 27 or key == ord('q'):
        cv2.destroyAllWindows()
        break
log_file.close()
```
